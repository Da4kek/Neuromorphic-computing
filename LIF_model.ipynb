```python
import snntorch as snn
from snntorch import spikeplot as splt
import torch
```

```python
import matplotlib.pyplot as plt 
from IPython.display import HTML
%matplotlib inline
```

```python
lif = snn.Leaky(beta=0.8)
```

```python
num_steps = 200
w = 0.8
x = torch.cat((torch.zeros(10), torch.ones(100)*w, torch.zeros(189)))
mem = torch.zeros(10)
spk = torch.zeros(10)
```

```python
mem_rec = []
spk_rec = []
for step in range(num_steps):
    spk,mem = lif(x[step],mem)
    mem_rec.append(mem)
    spk_rec.append(spk)

mem_rec = torch.stack(mem_rec)
spk_rec = torch.stack(spk_rec)
```

```python
from Utils import plot 
plot.Plotcurrentspike(x,mem_rec,spk_rec,thr_line=1,y_lim=1.0,title="leaky")
```
