```python
import snntorch as snn 
import torch
```

```python
import matplotlib.pyplot as plt 
import snntorch.spikeplot as splt
data = torch.Tensor([0,1,0,2,5,-20,-52,0,1,0])
plt.plot(data)
```

```python
from snntorch.spikegen import delta
spike_data = delta(data,threshold=4)
fig = plt.figure(facecolor="w",figsize=(8,1))
ax = fig.add_subplot(111)
splt.raster(spike_data,ax,c="black")
plt.xlim(0,len(data))
plt.yticks([])
```

```python
spike_data = delta(data,threshold=4,off_spike=True)
fig = plt.figure(facecolor="w",figsize=(8,1))
ax = fig.add_subplot(111)
splt.raster(spike_data,ax,c="black")
plt.xlim(0,len(data))
plt.yticks([])
```

```python
num_steps = 100
spike_prob = torch.rand((num_steps,28,28),dtype=torch.float)*0.5
spike_rand = delta(spike_prob)
```

```python
from IPython.display import HTML
fig,ax = plt.subplots()
anim = splt.animator(spike_rand,fig,ax)
HTML(anim.to_html5_video())
```

```python
fig = plt.figure(facecolor="w",figsize=(10,5))
ax = fig.add_subplot(111)
splt.raster(spike_rand[:,0].view(num_steps,-1),ax,s=25,c="black")
```
